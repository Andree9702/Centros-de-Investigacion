```python
import pandas as pd
import plotly.graph_objects as go # Import plotly.graph_objects
import plotly.express as px # Keep px for other potential uses, though not for heatmap directly

# Get the top N most frequent 'LÍNEA DE INVESTIGACIÓN' and 'CAMPO DETALLADO' categories
top_n = 10
top_investigacion = df['LÍNEA DE INVESTIGACIÓN'].value_counts().nlargest(top_n).index
top_campo_detallado = df['CAMPO DETALLADO'].value_counts().nlargest(top_n).index

# Filter the DataFrame to include only rows with these top categories
df_filtered = df[
    df['LÍNEA DE INVESTIGACIÓN'].isin(top_investigacion) &
    df['CAMPO DETALLADO'].isin(top_campo_detallado)
]

# Create a contingency table (crosstab) for the filtered data
contingency_table = pd.crosstab(
    df_filtered['LÍNEA DE INVESTIGACIÓN'],
    df_filtered['CAMPO DETALLADO']
)

# Create an interactive heatmap using Plotly Graph Objects
fig = go.Figure(data=go.Heatmap(
                   z=contingency_table.values,
                   x=contingency_table.columns,
                   y=contingency_table.index,
                   colorscale='YlGnBu')) # Changed colorscale to 'YlGnBu' for pastel effect

fig.update_layout(
    title=f'Correlación entre las Top {top_n} Líneas de Investigación y Campos Detallados',
    xaxis_title='Campo Detallado',
    yaxis_title='Línea de Investigación',
    xaxis_title_standoff=25,
    yaxis_title_standoff=25,
    height=700, # Adjust height for better viewing
    width=1000 # Adjust width for better viewing
)

fig.show()
```

```python
# Flatten the contingency table into a Series for easier sorting
contingency_series = contingency_table.stack()

# Sort the Series in descending order and get the top 5
top_5_combinations = contingency_series.nlargest(5)

if not top_5_combinations.empty:
    print("Las 5 combinaciones con el mayor número de producciones en el heatmap:")
    for (linea, campo), count in top_5_combinations.items():
        print(f"- Línea de Investigación: {linea}\n  Campo Detallado: {campo}\n  Número de Producciones: {count}\n")
else:
    print("No se encontraron combinaciones en la tabla de contingencia.")
```

ANALISIS FILTRADO DE LA PRODUCCION CIENTIFICA EXPERIMENTAL

```python
campos_a_excluir_keywords = [
    'Educación',
    'Economía',
    'Derecho',
    'Psicología',
    'Administración',
    'Mercadotecnia y publicidad',
    'Contabilidad y auditoría' # Added 'Contabilidad y auditoría'
]

# Create a boolean mask for rows to exclude
# Initialize with False (no rows excluded yet)
mask_to_exclude = pd.Series([False] * len(df), index=df.index)

# Check 'LÍNEA DE INVESTIGACIÓN' column
for keyword in campos_a_excluir_keywords:
    mask_to_exclude = mask_to_exclude | df['LÍNEA DE INVESTIGACIÓN'].str.contains(keyword, case=False, na=False)

# Check 'CAMPO DETALLADO' column
for keyword in campos_a_excluir_keywords:
    mask_to_exclude = mask_to_exclude | df['CAMPO DETALLADO'].str.contains(keyword, case=False, na=False)

# Filter the DataFrame: keep rows where mask_to_exclude is False
df_filtered_produccion = df[~mask_to_exclude]

print("DataFrame después de filtrar las líneas de investigación y campos detallados:\n")
print(df_filtered_produccion.head())
```

```python
num_rows, num_cols = df_filtered_produccion.shape
print(f"El DataFrame filtrado tiene {num_rows} filas y {num_cols} columnas.")
```

```python
import pandas as pd
import plotly.graph_objects as go # Import plotly.graph_objects
import plotly.express as px # Keep px for other potential uses, though not for heatmap directly

# Get the top N most frequent 'LÍNEA DE INVESTIGACIÓN' and 'CAMPO DETALLADO' categories
top_n = 5 # Changed top_n to 5 as requested by the user
top_investigacion_filtered = df_filtered_produccion['LÍNEA DE INVESTIGACIÓN'].value_counts().nlargest(top_n).index
top_campo_detallado_filtered = df_filtered_produccion['CAMPO DETALLADO'].value_counts().nlargest(top_n).index

# Filter the df_filtered_produccion DataFrame to include only rows with these top categories
df_filtered_for_heatmap = df_filtered_produccion[
    df_filtered_produccion['LÍNEA DE INVESTIGACIÓN'].isin(top_investigacion_filtered) &
    df_filtered_produccion['CAMPO DETALLADO'].isin(top_campo_detallado_filtered)
]

# Create a contingency table (crosstab) for the filtered data
contingency_table_filtered = pd.crosstab(
    df_filtered_for_heatmap['LÍNEA DE INVESTIGACIÓN'],
    df_filtered_for_heatmap['CAMPO DETALLADO']
)

# Print the actual number of rows and columns in the contingency table
print(f"Número de Líneas de Investigación en el heatmap: {len(contingency_table_filtered.index)}")
print(f"Número de Campos Detallados en el heatmap: {len(contingency_table_filtered.columns)}")

# Create an interactive heatmap using Plotly Graph Objects
fig = go.Figure(data=go.Heatmap(
                   z=contingency_table_filtered.values,
                   x=contingency_table_filtered.columns,
                   y=contingency_table_filtered.index,
                   colorscale='YlGnBu',
                   text=contingency_table_filtered.values, # Add text annotations
                   texttemplate="%{text}", # Format text to show the value
                   hoverinfo='text' # Ensure hoverinfo still works
                   ))

fig.update_layout(
    title=f'Correlación entre las Top {top_n} Líneas de Investigación y Campos Detallados (Filtrado)',
    xaxis_title='Campo Detallado',
    yaxis_title='Línea de Investigación',
    xaxis_title_standoff=25,
    yaxis_title_standoff=25,
    height=700, # Adjust height for better viewing
    width=1000 # Adjust width for better viewing
)

fig.show()
```

```python
# Flatten the contingency table into a Series for easier sorting
contingency_series = contingency_table_filtered.stack()

# Sort the Series in descending order and get the top 5
top_5_combinations = contingency_series.nlargest(5)

if not top_5_combinations.empty:
    print("Las 5 combinaciones con el mayor número de producciones en el heatmap (filtrado):\n")
    for (linea, campo), count in top_5_combinations.items():
        print(f"- Línea de Investigación: {linea}\n  Campo Detallado: {campo}\n  Número de Producciones: {count}\n")
else:
    print("No se encontraron combinaciones en la tabla de contingencia filtrada.")
```

```python
# Get the top 5 (Línea de Investigación, Campo Detallado) combinations from the filtered heatmap analysis
top_5_indices = top_5_combinations.index.tolist()

# Create a boolean mask to filter df_filtered_produccion
# Check if (LÍNEA DE INVESTIGACIÓN, CAMPO DETALLADO) in df_filtered_produccion matches any of the top 5 combinations
mask_for_top_5_productions = df_filtered_produccion.apply(
    lambda row: (row['LÍNEA DE INVESTIGACIÓN'], row['CAMPO DETALLADO']) in top_5_indices,
    axis=1
)

# Filter the DataFrame to get only these productions
df_top_5_productions = df_filtered_produccion[mask_for_top_5_productions]

# Calculate the value counts for 'CUARTIL' in this filtered DataFrame
cuartil_counts_top_5 = df_top_5_productions['CUARTIL'].value_counts()

print("Distribución de Cuartiles para las producciones de las 5 combinaciones principales:")
if not cuartil_counts_top_5.empty:
    print(cuartil_counts_top_5)
    print("\nEl Cuartil que más produce dentro de estas 5 combinaciones principales es: ")
    print(cuartil_counts_top_5.idxmax())
else:
    print("No se encontraron producciones que coincidan con las 5 combinaciones principales.")
```

```python
experimental_keywords = [
    'experimento', 'experimental', 'laboratorio', 'química', 'químico',
    'sustancias', 'materiales', 'reactivos', 'análisis', 'método',
    'pruebas', 'ensayo', 'equipo', 'instrumento', 'muestra', 'muestreo',
    'in vitro', 'in vivo', 'síntesis', 'purificación', 'secuenciación',
    'espectroscopia', 'cromato', 'microscopia', 'cultivo', 'medición',
    'protocolo', 'modelo experimental', 'validación'
]

# Function to check for experimental keywords in a title (case-insensitive)
def is_experimental(title):
    if isinstance(title, str):
        for keyword in experimental_keywords:
            if keyword in title.lower():
                return True
    return False

# Apply the function to the 'TITULO DE PRODUCCIÓN' column of the filtered DataFrame
df_filtered_produccion['CARACTER_EXPERIMENTAL'] = df_filtered_produccion['TITULO DE PRODUCCIÓN'].apply(is_experimental)

# Display the counts of experimental vs. non-experimental productions
experimental_counts = df_filtered_produccion['CARACTER_EXPERIMENTAL'].value_counts()

print("Análisis del carácter experimental de las producciones (basado en palabras clave en el título):\n")

if True in experimental_counts.index:
    print(f"Número de producciones experimentales: {experimental_counts[True]}")
else:
    print("Número de producciones experimentales: 0")

if False in experimental_counts.index:
    print(f"Número de producciones no experimentales: {experimental_counts[False]}")
else:
    print("Número de producciones no experimentales: 0")

print("\nPrimeras 5 producciones con el carácter experimental identificado:")
print(df_filtered_produccion[['TITULO DE PRODUCCIÓN', 'CARACTER_EXPERIMENTAL']].head())
```

```python
import pandas as pd
import plotly.graph_objects as go

# Filter the DataFrame to include only experimental productions
df_experimental = df_filtered_produccion[df_filtered_produccion['CARACTER_EXPERIMENTAL'] == True]

# Get the top N most frequent 'LÍNEA DE INVESTIGACIÓN' and 'CAMPO DETALLADO' categories
top_n_experimental = 10
top_investigacion_experimental = df_experimental['LÍNEA DE INVESTIGACIÓN'].value_counts().nlargest(top_n_experimental).index
top_campo_detallado_experimental = df_experimental['CAMPO DETALLADO'].value_counts().nlargest(top_n_experimental).index

# Filter the experimental DataFrame to include only rows with these top categories
df_experimental_for_heatmap = df_experimental[
    df_experimental['LÍNEA DE INVESTIGACIÓN'].isin(top_investigacion_experimental) &
    df_experimental['CAMPO DETALLADO'].isin(top_campo_detallado_experimental)
]

# Create a contingency table (crosstab) for the filtered experimental data
contingency_table_experimental = pd.crosstab(
    df_experimental_for_heatmap['LÍNEA DE INVESTIGACIÓN'],
    df_experimental_for_heatmap['CAMPO DETALLADO']
)

# Create an interactive heatmap using Plotly Graph Objects
fig = go.Figure(data=go.Heatmap(
                   z=contingency_table_experimental.values,
                   x=contingency_table_experimental.columns,
                   y=contingency_table_experimental.index,
                   colorscale='YlGnBu',
                   text=contingency_table_experimental.values, # Add text annotations
                   texttemplate="%{text}", # Format text to show the value
                   hoverinfo='text' # Ensure hoverinfo still works
                   ))

fig.update_layout(
    title=f'Correlación de Producciones Experimentales (Top {top_n_experimental} Líneas/Campos)',
    xaxis_title='Campo Detallado',
    yaxis_title='Línea de Investigación',
    xaxis_title_standoff=25,
    yaxis_title_standoff=25,
    height=700, # Adjust height for better viewing
    width=1000 # Adjust width for better viewing
)

fig.show()

# Flatten the contingency table into a Series for easier sorting
contingency_series_experimental = contingency_table_experimental.stack()

# Sort the Series in descending order and get the top 5 combinations
top_5_experimental_combinations = contingency_series_experimental.nlargest(5)

if not top_5_experimental_combinations.empty:
    print("\nLas 5 combinaciones de Línea de Investigación y Campo Detallado con el mayor número de producciones experimentales:\n")
    for (linea, campo), count in top_5_experimental_combinations.items():
        print(f"- Línea de Investigación: {linea}\n  Campo Detallado: {campo}\n  Número de Producciones Experimentales: {count}\n")
else:
    print("No se encontraron combinaciones significativas de producciones experimentales en la tabla de contingencia.")
```
